```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header = 1)
```

```python
data.info()
```

## Data Cleaning

```python
# missing values
data[data.isnull().any(axis = 1)]
```

The sataset is converted into two sets based on region from 122th index, we can make a new column based on the Region

1. "Bejaia Region Dataset"
2. "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
data.loc[:122, "Region"] = 0
data.loc[122:, "Region"] = 1
```

```python
data[['Region']] = data[['Region']].astype(int)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
# Remove the null values
data = data.dropna().reset_index(drop = True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.iloc[[122]]
```

```python
#remove the 122nd row
data.drop(122).reset_index(drop = True)
```

```python
data.iloc[[122]]
```

```python
data.columns
```

```python
# fix spaces in columns names
data.columns=data.columns.str.strip()
data.columns
```

#### Change the required columns as integer datatype

```python
data = data[data['month'] != 'month']
```

```python

data[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']]=data[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
data.info()
```

#### Changing the other columns to float data type

```python
objects = [features for features in data.columns if data[features].dtypes=='O']
```

```python
for i in objects:
    if i != 'Classes':
        data[i]=data[i].astype(float)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.head()
```

```python
## Let save cleaned dataset
data.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index = False)
```

## Exploratory Data Analysis

```python
# drop day, month , year
df = data.drop(['day','month','year'], axis = 1)
```

```python
df.head()
```

```python
# encoding of the categories in classes
df['Classes']=np.where(df['Classes']=='not fire', 0,1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
# plot dignity plot for all features
#plt.style.use('seaborn')
df.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# percentage for pie chart
percentage=df["Classes"].value_counts(normalize=True)*100
```

```python
#plotting pie chart
classlabels=["Fire", "Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```

```python
# correlation
df.corr()
```

```python
sns.heatmap(df.corr())
```
